```python
from amt_tools.tools.instrument import GuitarProfile
from guitar_transcription_continuous.datasets.GuitarSet import GuitarSetPlus
import os

class SmartGuitarSet(GuitarSetPlus):
    def __init__(self, *args, **kwargs):
        kwargs['profile'] = GuitarProfile()
        super().__init__(*args, **kwargs)

    def get_wav_path(self, track):
        base = self.base_dir

        if track.endswith("_comp"):
            track = track[:-5]
        elif track.endswith("_solo"):
            track = track[:-5]

        paths = [
            os.path.join(base, "audio", "mono_mic", f"{track}_comp_mic.wav"),
            os.path.join(base, "audio", "mono_mic", f"{track}_solo_mic.wav"),
            os.path.join(base, "audio_mono-mic", f"{track}_comp_mic.wav"),
        ]
        for path in paths:
            if os.path.exists(path):
                return path
        raise FileNotFoundError(f"Couldn't find any valid .wav file for track {track}")
```

```python
import pandas as pd

notes_raw = data["notes"]
all_notes = []

for string_entry in notes_raw:
    string_id = int(string_entry[0])
    pitches = string_entry[1][0]
    spans = string_entry[1][1]

    for pitch, (onset, offset) in zip(pitches, spans):
        all_notes.append({
            "string": string_id,
            "pitch": pitch,
            "onset": onset,
            "offset": offset,
            "duration": offset - onset
        })

df_notes = pd.DataFrame(all_notes)
df_notes.sort_values(by="onset", inplace=True)

# Tabloyu görüntüle
print(df_notes.head(10))
```

```python
import sys
sys.path.append("/content/amt-tools")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from amt_tools.tools.instrument import GuitarProfile

profile = GuitarProfile()

# .npz dosyasını yükle
track_path = "/content/amt-tools/generated/data/SmartGuitarSet/ground_truth/00_BN1-129-Eb_comp.npz"
data = np.load(track_path, allow_pickle=True)
notes_raw = data["notes"]

# 🔄 Nested yapıyı düzleştir
note_list = []
for string_index, (pitches, intervals) in notes_raw:
    for pitch, (onset, offset) in zip(pitches, intervals):
        duration = offset - onset
        note_list.append({
            "string": int(string_index),
            "pitch": pitch,
            "onset": onset,
            "offset": offset,
            "duration": duration
        })

# DataFrame oluştur
df_notes = pd.DataFrame(note_list)
df_notes["fret"] = df_notes.apply(lambda row: profile.get_fret(row["pitch"], int(row["string"])), axis=1)

# 🎨 Görselleştir
plt.figure(figsize=(12, 6))
scatter = plt.scatter(df_notes["onset"], df_notes["string"],
                      c=df_notes["fret"], cmap="viridis", s=60, edgecolors="k")

plt.colorbar(scatter, label="Fret")
plt.xlabel("Time (s)")
plt.ylabel("String")
plt.title("Guitar Tablature Visualization")
plt.gca().invert_yaxis()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## **CQT, HCQT ...**##

```python
import os
import sys
import numpy as np
import librosa
from tqdm import tqdm

# Gerekirse import path ekle
sys.path.append('/content/amt-tools')

# Feature modülleri
from amt_tools.features.cqt import CQT
from amt_tools.features.hcqt import HCQT
from amt_tools.features.mel import MelSpec
from amt_tools.features.stft import STFT
from amt_tools.features.power import SignalPower

# Yol ayarları
wav_folder = "/content/drive/MyDrive/Automatic Guitar Transcription/Data Set/GuitarSet/audio/mono_mic"
input_npz_folder = "/content/drive/MyDrive/Automatic Guitar Transcription/Data Set/GuitarSet/ground_truth_npz"
output_npz_folder = "/content/drive/MyDrive/Automatic Guitar Transcription/Data Set/GuitarSet/features_sr22050_all"
os.makedirs(output_npz_folder, exist_ok=True)

# Ortak örnekleme oranı (Tüm featurelar için 22050)
sr = 22050

# HCQT – Transformer için
hcqt_mod = HCQT(
    sample_rate=sr,
    hop_length=512,
    bins_per_octave=36,
    n_bins=144,
    harmonics=[0.5, 1, 2, 3, 4, 5],
    fmin=82.41,
    decibels=True
)

# CQT – CNN-CTC için
cqt_mod = CQT(
    sample_rate=sr,
    hop_length=512,
    fmin=32.70,  # C1
    n_bins=192,
    bins_per_octave=24,
    decibels=True
)

# Diğer modüller (TabCNN, CRNN, FretNet)
mel_mod = MelSpec(
    sample_rate=sr,
    hop_length=512,
    n_mels=229,
    n_fft=2048,
    decibels=True
)

power_mod = SignalPower(
    sample_rate=sr,
    hop_length=512,
    decibels=True
)

stft_mod = STFT(
    sample_rate=sr,
    hop_length=512,
    n_fft=2048,
    win_length=None,
    center=True,
    decibels=True
)

# .npz dosyalarını sırala
npz_files = sorted([f for f in os.listdir(input_npz_folder) if f.endswith(".npz")])

for npz_file in tqdm(npz_files, desc="Extracting features"):
    npz_name = os.path.splitext(npz_file)[0]
    input_npz_path = os.path.join(input_npz_folder, npz_file)
    output_npz_path = os.path.join(output_npz_folder, npz_file)

    wav_filename = npz_name + "_mic.wav"
    wav_path = os.path.join(wav_folder, wav_filename)

    if not os.path.exists(wav_path):
        print(f"❌ Missing WAV for {npz_file} → {wav_filename}")
        continue

    try:
        audio, _ = librosa.load(wav_path, sr=sr, mono=True)
    except Exception as e:
        print(f"⚠️ Failed to load audio for {npz_file}: {e}")
        continue


    try:
        data = dict(np.load(input_npz_path, allow_pickle=True))
    except Exception as e:
        print(f"⚠️ Could not load .npz file: {input_npz_path}: {e}")
        continue

    try:
        data["hcqt"] = hcqt_mod.process_audio(audio)
        data["cqt"] = cqt_mod.process_audio(audio)
        data["mel"] = mel_mod.process_audio(audio)
        data["stft"] = stft_mod.process_audio(audio)
        data["power"] = power_mod.process_audio(audio)
    except Exception as e:
        print(f"⚠️ Feature extraction error for {npz_file}: {e}")
        continue

    try:
        np.savez_compressed(output_npz_path, **data)
    except Exception as e:
        print(f"❌ Failed to save updated .npz: {output_npz_path}: {e}")
```
