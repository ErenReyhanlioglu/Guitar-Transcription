```python
import os

def scan_directory(base_dir, max_depth=2, max_files_preview=20, current_depth=0, prefix=""):
    if current_depth > max_depth:
        return

    try:
        items = sorted(os.listdir(base_dir))
    except Exception as e:
        print(f"{prefix}❌ Hata: {e}")
        return

    dirs = [d for d in items if os.path.isdir(os.path.join(base_dir, d))]
    files = [f for f in items if os.path.isfile(os.path.join(base_dir, f))]

    if current_depth == 0:
        print(f"\n📁 {base_dir} altında toplam {len(dirs)} klasör var.\n")

    indent = "  " * current_depth

    for d in dirs[:max_depth]:
        full_path = os.path.join(base_dir, d)
        sub_items = os.listdir(full_path)
        print(f"{indent}📂 {d} ({len(sub_items)} içerik)")
        scan_directory(full_path, max_depth, max_files_preview, current_depth + 1)

    for f in files[:max_files_preview]:
        print(f"{indent}  └── 📄 {f}")
    if len(files) > max_files_preview:
        print(f"{indent}  └── ... {len(files) - max_files_preview} diğer dosya")
```

```python
import os

# Klasör yolları
jams_path = "/content/drive/MyDrive/Automatic Guitar Transcription/Data Set/SynthTab/all_jams/outall"
midi_folder = "/content/drive/MyDrive/Automatic Guitar Transcription/Data Set/SynthTab/luthier_pick/part_1/Alice In Chains - Got Me Wrong (Unplugged) - gp4__2 - Acoustic Steel Guitar__midi"
paths_folder = "/content/drive/MyDrive/Automatic Guitar Transcription/Data Set/SynthTab/all_paths/allpaths_acoustic"

# Klasörleri listeleyen yardımcı fonksiyon
def list_folder(path, name):
    if os.path.exists(path) and os.path.isdir(path):
        files = os.listdir(path)
        print(f"📁 {name} → Toplam {len(files)} dosya")
        print("🧾 İlk 15 dosya:", files[:15])
    else:
        print(f"⚠️ {name} klasörü bulunamadı veya bir dizin değil!")

# İnceleme
list_folder(jams_path, "JAMS_OUTALL")
list_folder(midi_folder, "MIDI_FOLDER (Got Me Wrong)")
list_folder(paths_folder, "ALL_PATHS")
```
