```python
import os

def scan_directory(base_dir, max_depth=2, max_files_preview=20, current_depth=0, prefix=""):
    if current_depth > max_depth:
        return

    try:
        items = sorted(os.listdir(base_dir))
    except Exception as e:
        print(f"{prefix}âŒ Hata: {e}")
        return

    dirs = [d for d in items if os.path.isdir(os.path.join(base_dir, d))]
    files = [f for f in items if os.path.isfile(os.path.join(base_dir, f))]

    if current_depth == 0:
        print(f"\nğŸ“ {base_dir} altÄ±nda toplam {len(dirs)} klasÃ¶r var.\n")

    indent = "  " * current_depth

    for d in dirs[:max_depth]:
        full_path = os.path.join(base_dir, d)
        sub_items = os.listdir(full_path)
        print(f"{indent}ğŸ“‚ {d} ({len(sub_items)} iÃ§erik)")
        scan_directory(full_path, max_depth, max_files_preview, current_depth + 1)

    for f in files[:max_files_preview]:
        print(f"{indent}  â””â”€â”€ ğŸ“„ {f}")
    if len(files) > max_files_preview:
        print(f"{indent}  â””â”€â”€ ... {len(files) - max_files_preview} diÄŸer dosya")
```
